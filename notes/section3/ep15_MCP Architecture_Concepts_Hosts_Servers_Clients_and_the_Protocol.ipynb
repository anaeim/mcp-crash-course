### Why MCP?
MCP helps you build agents and complex workflows on top of LLMs. LLMs frequently need to integrate with data and tools, and MCP provides:

- A growing list of pre-built integrations that your LLM can directly plug into
- The flexibility to switch between LLM providers and vendors
- Best practices for securing your data within your infrastructure

### MCP server
servers are the component that will expose those resources, those tools, those prompts.<br> 
they will be the proxy, the gateway that will expose that functionality.

### MCP client
MCP client sits inside the MCP hosts.<br>
MCP client is going to be able to interact and to talk with the MCP servers. And they're going to do this via the MCP protocol. So protocol is language that they're going to interact with.

It's important to note that there's a one-to-one connection between an MCP client and an MCP server. This means a single MCP client cannot connect to multiple MCP servers. If you want an MCP host to connect to multiple MCP servers, you'll need to run multiple MCP clients on that host.
